```python
direccion=input("Direccion IP : ")
```

```python
Sufijo=int(input("Numero de Sufijo SIDR : "))
```

```python
direccion2=direccion.split(".")
direccion2
```

```python
def host():
    bist_host=32
    porcion=8
    bists=bist_host
    resultado=[]
    bists=bists-Sufijo
    resultado.append(bists)
    numero_hosts=(2**bists)
    resultado.append(numero_hosts)
    bists_sub_red=porcion-bists
    resultado.append(bists_sub_red)
    return resultado
```

```python
def muestraDatos():
    datos=host()
    print("\nDireccion Origen = {} \nNumero de CIDR = {} \nPorcion Bits Host = {} \nPorcion Sub Red = {} \nNumero de Host = {}".format(direccion,Sufijo,datos[0],datos[2],datos[1]-2))
```

```python
def ipDisposnibles():
    datos=host()
    c=0
    for i in range((2**datos[2])):
        print("---------------------------------------------------------------\nSub Red : {}\n".format(i+1))
        for j in range (datos[1]):
            if j==0:
                print("Id. de Red : "+direccion+".",c)
            elif j == datos[1]-1:
                print("Broadcast   : "+direccion+".",c)
            else:
                print("Host ",j,"   : "+direccion+".",c)
            c+=1
```

```python
muestraDatos()
ipDisposnibles()
```
